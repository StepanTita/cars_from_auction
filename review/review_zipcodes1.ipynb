## Merge data

```python
def merge_data(data, zipcodes):
    zipcodes = zipcodes.drop_duplicates(keep='last', subset=['zipcode'])
    mdata = data.merge(zipcodes, on="zipcode", how="left") 
    return mdata
```

```python
mdata = merge_data(train_data, zip_data)
```

```python
def is_high(row):
    clone = row.copy()
    clone[row.abs() > 0.3] = '#ff937d'
    clone[row.abs() > 0.5] = '#f56942'
    clone[row.abs() > 0.6] = '#ff795e'
    clone[(row.abs() <= 0.3) | (row.abs() == 1.0)] = ''
    return [f'background-color: {cell}' for cell in clone]
```

```python
mdata.corr().style.apply(is_high)
```
